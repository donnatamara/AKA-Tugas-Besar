```python
import time
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate

# List untuk menyimpan nilai n dan waktu eksekusi
n_values = []
recursive_times = []
iterative_times = []

# Fungsi Merge Sort Rekursif
def merge_sort_recursive(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort_recursive(left)
        merge_sort_recursive(right)

        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i][1] < right[j][1]:  # Sort berdasarkan tahun (Year)
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

# Fungsi Merge Sort Iteratif
def merge_sort_iterative(arr):
    n = len(arr)
    width = 1
    while width < n:
        l = 0
        while l < n:
            r = min(l + (width * 2 - 1), n - 1)
            mid = min(l + width - 1, n - 1)
            left = arr[l:mid + 1]
            right = arr[mid + 1:r + 1]
            i = j = k = 0
            while i < len(left) and j < len(right):
                if left[i][1] < right[j][1]:  # Sort berdasarkan tahun (Year)
                    arr[l + k] = left[i]
                    i += 1
                else:
                    arr[l + k] = right[j]
                    j += 1
                k += 1
            while i < len(left):
                arr[l + k] = left[i]
                i += 1
                k += 1
            while j < len(right):
                arr[l + k] = right[j]
                j += 1
                k += 1
            l += width * 2
        width *= 2

# Fungsi mencetak tabel eksekusi
def print_execution_table():
    print(tabulate(
        zip(n_values, recursive_times, iterative_times),
        headers=["n", "Recursive Time (s)", "Iterative Time (s)"],
        tablefmt="grid"
    ))

# Fungsi memperbarui grafik
def update_graph():
    plt.clf()  # Bersihkan grafik sebelumnya
    plt.plot(n_values, recursive_times, 'o-', label="Recursive Time")
    plt.plot(n_values, iterative_times, 'o-', label="Iterative Time")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Execution Time (seconds)")
    plt.title("Comparison of Recursive vs Iterative Merge Sort")
    plt.legend()
    plt.grid()
    plt.pause(0.01)  # Perbarui grafik

# Program utama
if __name__ == "__main__":
    print("Perbandingan Algoritma Merge Sort Rekursif dan Iteratif")

    df = df[['Title', 'Year', 'Rating']].dropna()  # Ambil kolom relevan dan hilangkan NaN

    while True:
        try:
            n = int(input("Masukkan jumlah data film (n) (atau ketik -1 untuk keluar): "))
            if n == -1:
                print("Program selesai. Terima kasih!")
                break
            if n <= 0:
                print("Masukkan nilai n yang positif!")
                continue

            # Ambil n data dari dataset
            sample_data = df.head(n).values.tolist()

            # Ukur waktu eksekusi Merge Sort Rekursif
            recursive_data = sample_data.copy()
            start_time = time.time()
            merge_sort_recursive(recursive_data)
            recursive_time = time.time() - start_time
            recursive_times.append(recursive_time)

            # Ukur waktu eksekusi Merge Sort Iteratif
            iterative_data = sample_data.copy()
            start_time = time.time()
            merge_sort_iterative(iterative_data)
            iterative_time = time.time() - start_time
            iterative_times.append(iterative_time)

            # Simpan nilai n
            n_values.append(n)

            # Cetak tabel dan perbarui grafik
            print_execution_table()
            update_graph()

            # Menampilkan hasil pengurutan berdasarkan Tahun Rilis
            print("\nHasil Pengurutan Berdasarkan Tahun Rilis:")
            sorted_data = recursive_data  # Bisa menggunakan iterative_data juga, hasilnya sama
            for movie in sorted_data:
                print(f"Title: {movie[0]}, Year: {movie[1]}, Rating: {movie[2]}")

        except ValueError:
            print("Masukkan nilai n yang valid!")

    # Menampilkan grafik terakhir
    plt.show()
```
